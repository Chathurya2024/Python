```python
import tkinter as tk
from tkinter import ttk
import random
import pandas as pd

# Creating a class named EastCoastNames
class EastCoastNames:
    def __init__(self, filename):
        self.data = pd.read_csv(filename)
        
    def first_name(self, state=None, gender=None):
        if state is None:
            state = random.choice(self.data["state"])
        else:
            state = state
        
        if gender is None:
            gender = random.choice(["m", "f"])
        else:
            gender = gender
        
        new_data = self.data[(self.data["state"] == state) & (self.data["gender"] == gender)].reset_index(drop=True)
        return random.choice(new_data["fname"])
    
    def last_name(self, state=None, gender=None):
        if state is not None and gender is not None:
            new_data = self.data[(self.data["state"] == state) & (self.data["gender"] == gender)].reset_index(drop=True)
        else:
            state = random.choice(self.data["state"])
            gender = random.choice(["m", "f"])
            new_data = self.data[(self.data["state"] == state) & (self.data["gender"] == gender)].reset_index(drop=True)
            
        return random.choice(new_data["lname"])
    
    def full_name(self, state=None, gender=None):
        firstname = self.first_name(state, gender)
        lastname = self.last_name(state, gender)
        return f"{firstname} {lastname}"

def generate_full_name():
    state = state_entry.get().lower()
    gender = gender_entry.get().lower()
    if state == "":
        state = None
    if gender == "":
        gender = None
    fullname = ecn.full_name(state, gender)
    result_label.configure(text=f"The generator has returned: {fullname}")


root = tk.Tk()
root.title("Random Name Generator")

# Create frame for inputs
frame = ttk.Frame(root, padding="10")
frame.grid(row=0, column=0)

# Create labels and entries for state and gender inputs
state_label = ttk.Label(frame, text="Enter the state name (or press enter for random):")
state_label.grid(row=0, column=0, sticky="w")
state_entry = ttk.Entry(frame)
state_entry.grid(row=0, column=1)

gender_label = ttk.Label(frame, text="Gender of the name (m/f) (or press enter for random):")
gender_label.grid(row=1, column=0, sticky="w")
gender_entry = ttk.Entry(frame)
gender_entry.grid(row=1, column=1)

# Create button to generate full name
generate_button = ttk.Button(frame, text="Generate Full Name", command=generate_full_name)
generate_button.grid(row=2, column=0, columnspan=2, pady=(10, 0))

# Create label to display generated full name
result_label = ttk.Label(frame, text="")
result_label.grid(row=3, column=0, columnspan=2, pady=(10, 0))


# Initialize EastCoastNames object
filename = "State_names.csv"
ecn = EastCoastNames(filename)

# Run GUI event loop
root.mainloop()
```
