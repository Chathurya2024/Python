```python
import matplotlib.pyplot as plt
import pandas as pd

# Read the metadata from the saved file (JSON or CSV)
data = pd.read_csv('papers.csv')  # Replace with the actual file path if different

# Filter the DataFrame for each primary category
databases_df = data[data['primary_category'] == 'cs.DB']
graphics_df = data[data['primary_category'] == 'cs.GR']
robotics_df = data[data['primary_category'] == 'cs.RO']
emerging_df = data[data['primary_category'] == 'cs.ET']

# Count the number of papers uploaded by year for each category
databases_count = databases_df['year'].value_counts().sort_index()
graphics_count = graphics_df['year'].value_counts().sort_index()
robotics_count = robotics_df['year'].value_counts().sort_index()
emerging_count = emerging_df['year'].value_counts().sort_index()

# Create a bar graph
years = range(2018, 2023)
bar_width = 0.2

plt.figure(figsize=(10, 6))
# Creating a new figure with a specific size
plt.bar(years, databases_count, width=bar_width, label='Databases')
plt.bar([year + bar_width for year in years], graphics_count, width=bar_width, label='Graphics')
plt.bar([year + 2 * bar_width for year in years], robotics_count, width=bar_width, label='Robotics')
plt.bar([year + 3 * bar_width for year in years], emerging_count, width=bar_width, label='Emerging Technologies')

plt.xlabel('Year')
plt.ylabel('Number of Papers')
plt.title('Number of Papers Uploaded by Year and Category')
plt.xticks([year + 1.5 * bar_width for year in years], years)
plt.legend()

plt.show()
```
