<b>Question 2:</b> [<span style="color: red;">10 points</span>]:<br>
The file <code>airtravel.csv</code> shows monthly transatlantic airtravel, in thousands of passengers, for 1958–1960. There are 4 fields, “Month”, “1958”, “1959” and “1960” and 12 records, “JAN” through “DEC”.<br>
Your task is to create a bargraph with the X-axis for the months and Y-axis for the number of passengers. Each month should have 3 bars for the three years in question.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Load CSV file into DataFrame
data = pd.read_csv('airtravel.csv')

length = np.arange(len(data["Month"]))


# Plot the bars for each year
plt.bar(length + 0 , data["1958"], width=0.2, color= '#bcbddc', label='1958')
plt.bar(length + 0.2, data["1959"], width=0.2, color= '#928db8', label='1959')
plt.bar(length + 2*0.2, data["1960"], width=0.2, color='#645e8f', label='1960')

# #756bb1
# Set labels and title
plt.xlabel('Month')
plt.ylabel('Passengers (thousands)')
plt.title('Monthly Transatlantic Air Travel (1958-1960)')

# Set x-axis tick positions and labels
plt.xticks(length, data["Month"])

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

<b>Question 3:</b> [<span style="color: red;">10 points</span>]:<br>
The file <code>al_east_2012.csv</code> has the information about the teams in the AL East League (baseball), specifically the payroll (in millions) and number of wins. Create two pie charts where it shows each team's percentage share of (i) payroll and (ii) wins respectively. "Explode" the slice that is the largest by 0.1.

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
import pandas as pd
import matplotlib.pyplot as plt

# Load data from CSV file
AL_east = pd.read_csv('al_east_2012.csv')
AL_east
```

```python
# Sort the data by 'Payroll (in millions)' column in descending order
AL_east = AL_east.sort_values(by='Payroll (in millions)', ascending=False)

# Create a list of zeros with the same length as the data
explode = [0] * len(AL_east)

# Set the explode value for the slice with the maximum value
explode[0] = 0.1  # You can adjust the explode value as needed

# Create the pie chart with ordered slices plotted counter-clockwise
plt.pie(AL_east['Payroll (in millions)'], labels=AL_east['Team'], autopct='%1.1f%%', explode=explode, shadow=True, startangle=90, counterclock=True)

plt.title('AL East 2012 Payroll Distribution with Ordered Slices')
plt.show()
```

```python
# Sort the data by 'Wins' column in descending order
AL_east = AL_east.sort_values(by='Wins', ascending=False)

# Create a list of zeros with the same length as the data
explode = [0] * len(AL_east)

explode[0] = 0.1 

# Create the pie chart with ordered slices plotted counter-clockwise
plt.pie(AL_east['Wins'], labels=AL_east['Team'], autopct='%1.1f%%', explode=explode, shadow=True, startangle=90, counterclock=True)

plt.title('AL East 2012 Wins')
plt.show()
```

The file <code>hw_25000.csv</code> contains height and weight for 25000 individuals. Each record includes 3 values: index, height (inches), weight (pounds)<br>
Generate two separate histograms (separate figures) for the height as well as weight data.

```python
#Your answer goes here
data = pd.read_csv('hw_25000.csv')
data
```

```python
print(data.columns)
```

```python
fig, ax = plt.subplots(1,1,sharey=True, sharex=True, tight_layout=True)
#histogram
ax.hist(data["Height (Inches)"],bins=50,label="Height (Inches)")

fig.set_size_inches(4*1.45, 3*1.45)
fig.set_tight_layout(True)
fig.text(0.015, 0.5, 'Number of individuals', ha='center', va='center', rotation='vertical')
ax.legend(loc='upper left')

ax.set_xlabel('Height (Inches)')
ax.set_ylabel('')

plt.show()
```

```python
fig, ax = plt.subplots(1,1,sharey=True, sharex=True, tight_layout=True)

ax.hist(data[" Weight (Pounds)"],bins=50,label=" Weight (Pounds)")

fig.set_size_inches(4*1.45, 3*1.45)
fig.set_tight_layout(True)
fig.text(0.015, 0.5, 'Number of individuals', ha='center', va='center', rotation='vertical')
ax.legend(loc='upper left')

ax.set_xlabel(' Weight (Pounds)')
ax.set_ylabel('')

plt.show()
```
