```python
#Importing 'names' module.
import names
```

(i). Print your list and make sure that you indeed have 10 names in your lists.

```python
#We first create empty list.
fname = []
#We use 'for' loop to repeat the process for 'n' number of times.
#range(10) = [0,1,2,....,9]
for i in range(10):
    #we get random names using get_first_name function from name module.
    first_names = names.get_first_name()
    #We append all elements together.
    fname.append(first_names)
#We are using 'f-string' to embed expressions inside a string.
print(f" List of first 10 First names: {fname}")

#Similarly we do this for the last name. Here we change the function to get_last_name.
lname = []
for j in range(10):
    last_names = names.get_last_name()
    lname.append(last_names)
print(f" List of first 10 Last names: {lname}")
```

(ii). Remove the last "first-name" from your "fname" list.

```python
#Pop function removes the last item by default, if index is not passed.
fname.pop()
print(f" First names: {fname}")
#print(len(fname))
```

(iii). Insert a new last-name between the second-last and last items in the "lname" list.

```python
#insert function is used to add any element in any particular position.
lname.insert(9,names.get_last_name())
print(lname)
```

(iv). Add a first-name to the end of "fname" and remove the first item in the same list.

```python
#Adding first_name at the end of fname
fname.insert(10,names.get_first_name())
print(fname)

#Here we are giving index in pop to remove first name.
fname.pop(0)
print(fname)
```

(v). Generate three random first names, print them, and then add three first names at the end of your "fname" list using the "append" command.

```python
#Generating three random first names and adding them to the list 'fname'.
for i in range(0,3):
    three_random_names = names.get_first_name()
    fname.append(three_random_names)
print(fname)
```

```python
import random
# Importing randint function from the random module
from random import randint
```

(i). 10 random birth-years between 1950 and 2022.

```python
#We are printing random years using 'randint'
byears = []
for i in range(0,10):
    years = randint(1950, 2022)
    byears.append(years)
print(byears)
```

(ii). 10 random birth-months between 1 and 12.

```python
bmonths = []
for i in range(0,10):
    months = randint(1,12)
    bmonths.append(months)
print(bmonths)
```

(iii). 10 random birth-days between 1 and 31.

```python
bdays = []
#for months - leap year is used.
for i in range(10):
    if [1,3,5,7,8,10,12] in bmonths:
        days = randint(1,31)
    elif bmonths == 2:
        if byears % 4 == 0:
            days = randint(1,29)
        else:
            days = randint(1,28)
    else:
        days = randint(1,30)
    bdays.append(days)
print(bdays)
```

```python
#Printing birth date, month and year all together.
for i in range(10):
    print(f"{bdays[i]} - {bmonths[i]} - {byears[i]}")
```

(iv) Create 10 random Credit-Scores between 100 and 800. Create a list of "CreditApproval" with "Yes" if the credit-score is higher than 650, "No" otherwise. Make sure that someone can't get a credit if they are 17 years old or younger, even if their credit-score is higher than 650. (You can find age by just subtracting their birth year from 2023)

```python
#Calculating age from birth year
age = byears
for i in range(len(age)):
    age[i] = 2023 - age[i]
print(age)
```
